```python
state=df.groupby("State")[["Sales","Profit"]].sum().sort_values(by="Sales",ascending=False)
plt.figure(figsize=(20,20))
state[:25].plot(kind="bar",color=["k","yellow"],edgecolor="#000000")
plt.title("profit or loss and sale of top 25 sales")
plt.xlabel("States")
plt.ylabel("total profit / loss and sales")
plt.grid(True)
state[25:].plot(kind="bar",color=["k","yellow"],edgecolor="#000000")
plt.title("profit or loss and sale of least economic states")
plt.xlabel("States")
plt.ylabel("total profit / loss and sales")
plt.grid(True)
```

```python
df.groupby('Sub-Category')['Profit','Sales'].sum().plot.bar(color=['b','k'])
```

```python
pd.DataFrame(df.groupby('State').sum())['Profit'].sort_values(ascending=True)
```

```python
pd.DataFrame(df.groupby('State').sum())['Discount'].sort_values(ascending=True)
```

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit',data=df,color='r',label='Discount')
plt.legend()
```

from the above study.when the profit of a state are compared with the discount provided in each state. the states which allowed more discount. went in loss and the above graph we conclude that THE PROFIT DECREASES WITH THE INCREASE IN DISCOUNT.
